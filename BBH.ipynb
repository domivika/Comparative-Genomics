```python
# load in blastp outputs - forward and reverse alignments
# forward = acetivibrio vs methylosinus
# reverse = methylosinus vs acetivibrio
forward_hits = pd.read_csv(os.path.join(work_dir, 'forward_blastp'), sep='\t', header=None)
reverse_hits = pd.read_csv(os.path.join(work_dir, 'reverse_blastp'), sep='\t', header=None)
```

```python
reverse_hits
```

```python
# merge forward and reverse arrays
array = np.vstack((forward_hits, reverse_hits))
print(array)
```

```python
# convert array to pandas dataframe
df = pd.DataFrame(array)
df
```

```python
# name headers
header = ["query", "subject", "identity", "allignment_length",
          "mismatches", "gap_openings", "start_query",
          "end_query", "start_subject", "end_subject",
          "E-value", "bitscore"]
```

```python
# add headers to the dataframe
df.columns = header
df
```

```python
# keep only unique rows - based on query and subject columns
# if duplicates - choose the row with he highest bitscore
def keep_max_bitscore(group):
    # return the row with the maximum value of bitscore
    return group.iloc[group['bitscore'].argmax()]

# convert bitscore column to a numerical data type
df['bitscore'] = df['bitscore'].astype(int)

df_deduplicated = df.groupby(['query', 'subject']).apply(keep_max_bitscore)
```

```python
df_deduplicated
```

```python
# keep only Reciprocal Best Hits (RBHs)
def keep_RBHs(df,col1, col2):
    new_df = df[(df[col1].isin(df[col2])) & (df[col2].isin(df[col1]))]
    return new_df

result = keep_RBHs(df_deduplicated,'query', 'subject')
result
```
